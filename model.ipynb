```python
import re
import emoji
import nltk
from nltk.corpus import stopwords
from unidecode import unidecode
import pandas as pd

nltk.download('stopwords')

# Load your dataset
df = pd.read_csv("final_dataset_30k_fixed.csv")
df = df.dropna(subset=['sentence'])

# --- Hinglish normalization ---
hinglish_dict = {
    "acha": "achha", "accha": "achha", "achaa": "achha",
    "thik": "theek", "mast": "great", "bohot": "bahut",
    "bahot": "bahut", "nahi": "nahin", "nhi": "nahin",
    "pls": "please", "plz": "please", "frnd": "friend",
    "gud": "good", "luv": "love", "tera": "your", "meri": "my",
    "bht": "bahut", "bahut": "bahut", "kya": "what", "kyu": "why"
}

# --- Emoji to sentiment mapping ---
emoji_dict = {
    "😀": "happy", "😃": "happy", "😄": "happy", "😁": "happy",
    "😆": "laugh", "😂": "laugh", "🤣": "laugh", "😊": "happy",
    "😍": "love", "🥰": "love", "😘": "love", "😎": "cool",
    "😢": "sad", "😭": "cry", "😡": "angry", "😠": "angry",
    "❤️": "love", "💔": "broken_heart", "🔥": "amazing",
    "👍": "good", "👎": "bad", "🙏": "thankful", "👏": "applause"
}

stop_words = set(stopwords.words('english')) - {'not', 'no', 'nor'}

def emoji_to_word(text):
    for emj, meaning in emoji_dict.items():
        text = text.replace(emj, f" {meaning} ")
    return text

def clean_hinglish(text):
    if not isinstance(text, str):
        return ""
    text = text.lower()
    text = emoji_to_word(text)
    text = re.sub(r"http\S+|www\S+|@\S+|#\S+", " ", text)
    text = re.sub(r"[^a-zA-Z!? ]", " ", text)
    for key, val in hinglish_dict.items():
        text = re.sub(r"\b" + re.escape(key) + r"\b", val, text)
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)
    tokens = [word for word in text.split() if word not in stop_words]
    text = " ".join(tokens)
    text = unidecode(text)
    text = re.sub(r"\s+", " ", text).strip()
    return text

# Apply cleaning
df['clean_text'] = df['sentence'].apply(clean_hinglish)

# Drop very short/gibberish lines
df = df[df['clean_text'].str.split().str.len() > 2]

# Save cleaned version
df.to_csv("final_dataset_30k_preprocessed.csv", index=False)

print("✅ Deep preprocessing + emoji mapping complete")
print("New shape:", df.shape)
print(df[['sentence', 'clean_text']].sample(5, random_state=42))
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Load your dataset
df = pd.read_csv("/Users/aashigarg/Downloads/cleaned_dataset_updated11.csv").dropna(subset=["clean_text"])

# Map labels to numeric values
label_map = {"negative": 0, "neutral": 1, "positive": 2}
df["label"] = df["Sentiment"].map(label_map)

# Verify mapping
print(df["label"].value_counts())
```

```python
import re

def preprocess_text(text):
    if not isinstance(text, str):
        return ""
    # Remove URLs, mentions, hashtags (keep the word after '#')
    text = re.sub(r"http\S+|www\S+|@\w+", "", text)
    text = re.sub(r"#", "", text)

    # Remove unwanted symbols but keep Hindi (Devanagari) + English words + emojis
    text = re.sub(r"[^a-zA-Z0-9\s\u0900-\u097F🙂😂😭😡😍😢😃😅🙄😉🙃😎🤔🤩❤️💕💔🔥🙌👏💯✨⭐🌟🙏]", " ", text)

    # Normalize elongated letters: e.g. "goooood" → "good"
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)

    # Replace multiple spaces with one
    text = re.sub(r'\s+', ' ', text).strip()

    return text

# Apply preprocessing
df["clean_text"] = df["clean_text"].apply(preprocess_text)

# Verify
df.head(10)
```

```python
train_texts, temp_texts, train_labels, temp_labels = train_test_split(
    df["clean_text"],
    df["label"],
    test_size=0.2,
    stratify=df["label"],
    random_state=42
)

val_texts, test_texts, val_labels, test_labels = train_test_split(
    temp_texts,
    temp_labels,
    test_size=0.5,
    stratify=temp_labels,
    random_state=42
)
```

```python
from transformers import BertTokenizer

tokenizer = BertTokenizer.from_pretrained("bert-base-multilingual-cased")

def tokenize_function(texts):
    return tokenizer(
        list(texts),
        padding="max_length",
        truncation=True,
        max_length=128,
        return_tensors="pt"
    )

train_encodings = tokenize_function(train_texts)
val_encodings = tokenize_function(val_texts)
test_encodings = tokenize_function(test_texts)
```

```python
import torch

class SentimentDataset(torch.utils.data.Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __getitem__(self, idx):
        item = {key: val[idx] for key, val in self.encodings.items()}
        item["labels"] = torch.tensor(self.labels.iloc[idx])
        return item

    def __len__(self):
        return len(self.labels)

train_dataset = SentimentDataset(train_encodings, train_labels)
val_dataset = SentimentDataset(val_encodings, val_labels)
test_dataset = SentimentDataset(test_encodings, test_labels)
```

```python
from transformers import BertForSequenceClassification, Trainer, TrainingArguments
import numpy as np
from sklearn.metrics import accuracy_score, f1_score

model = BertForSequenceClassification.from_pretrained(
    "bert-base-multilingual-cased",
    num_labels=3
)

from transformers import TrainingArguments

training_args = TrainingArguments(
    output_dir="./results",
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=32,
    num_train_epochs=3,
    weight_decay=0.01,
    logging_dir='./logs',
    logging_steps=100,       # log progress every 100 steps
    save_total_limit=1,      # keep only last checkpoint
)

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    preds = np.argmax(logits, axis=1)
    return {
        "accuracy": accuracy_score(labels, preds),
        "macro_f1": f1_score(labels, preds, average="macro")
    }

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    compute_metrics=compute_metrics
)

trainer.train()
```
